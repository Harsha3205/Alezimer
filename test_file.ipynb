```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from tensorflow.keras.preprocessing.image import load_img,img_to_array
from tensorflow.keras.models import load_model
```

```python
class_labels={0:'MildDemented', 1:'NonDemented', 2:'VeryMildDemented'}
```

```python
model=load_model(filepath="models/Convolutional_Neural_Network.h5")
```

```python
filepath="ValidationImages/mild_demented/milddemented (1).jpg"
```

```python
 image=load_img(filepath, grayscale=True, color_mode='grayscale', target_size=(128,128))
```

```python
plt.imshow(image,cmap='gray')
plt.axis("off")
plt.show()
```

```python
image=img_to_array(image)
image=image/255.0
image=np.expand_dims(image,axis=0)
```

```python
model_pred=model.predict(image)
print(f"model predicted class is {np.argmax(model_pred[0])}")
print(f"model predicted label is {class_labels[np.argmax(model_pred[0])]}")
```
